```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense,Flatten,BatchNormalization,Dropout
from keras.optimizers import Adam
from keras.regularizers import L1,L2
from keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score

```

```python
data = pd.read_csv("telecom_churn.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data["Churn"].value_counts()
```

```python
x = data.drop("Churn",axis =1)
```

```python
y = data["Churn"]
```

```python
smote = SMOTE()
```

```python
x,y = smote.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
data.corr()*100
```

```python

num_feature = ["AccountWeeks","CustServCalls", "DayMins", "DayCalls", "MonthlyCharge", "OverageFee", "RoamMins"]
```

```python
scaler = StandardScaler()

for num in num_feature:
    x[num] = scaler.fit_transform(x[[num]])
```

```python
x
```

```python

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x_test
```

```python
model = Sequential(
    [ 
    (Dense(units =64, activation = "relu",input_shape=(x_train.shape[1],1),kernel_regularizer = L2(l2=0.01))),
    # (BatchNormalization()),
    # (Dropout(0.2)),
    (Dense(units =64//2, activation = "relu")),
    (BatchNormalization()),
    (Dense(units =16, activation = "relu")),
    (Dense(units =8, activation = "relu")),
    (Flatten()),
    (Dense(units =1, activation = "sigmoid"))
    ]
)
```

```python
model.compile(optimizer = Adam(learning_rate = 0.001),loss = "binary_crossentropy",metrics = ["accuracy"])
```

```python
x_train.shape
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,epochs = 20,batch_size =64,validation_split = 0.2,callbacks = EarlyStopping())
```

```python
y_pred1 = model.predict(x_train)
pred_data1 =[]
for i in y_pred1:
    if i[0]>0.5:
        pred_data1.append(1)
    else:
        pred_data1.append(0)
```

```python
y_pred = model.predict(x_test)
pred_data =[]
for i in y_pred:
    if i[0]>0.5:
        pred_data.append(1)
    else:
        pred_data.append(0)  
```

```python
pred_data
```

```python
accuracy_score(y_test,pred_data)*100
```

```python
accuracy_score(y_train,pred_data1)*100
```
